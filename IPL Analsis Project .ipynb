```python
#getting the frequency of result ccolumn 
ipl["result"].value_counts()
```

```python

#storing into results_counts
result_counts = ipl["result"].value_counts()
```

```python
# Create a bar chart
plt.figure(figsize=(8, 5))
result_counts.plot(kind='bar', color='skyblue')
plt.title('Result Counts')
plt.xlabel('Result')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
sns.set_style("whitegrid")

# Create a bar plot using Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x=result_counts.index, y=result_counts.values, palette="Blues_d")
plt.title('Result Counts')
plt.xlabel('Result')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Create a bar plot using Seaborn
plt.figure(figsize=(8, 5))
ax = sns.barplot(x=result_counts.index, y=result_counts.values, palette="Blues_d")

# Add value labels on the bars
for i, value in enumerate(result_counts.values):
    ax.text(i, value, str(value), ha='center', va='bottom', fontsize=10)

plt.title('Result Counts')
plt.xlabel('Result')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Extracting the records where team won batting first 
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
#looking at head
batting_first.head()
```

```python
batting_first.shape
```

```python
#making the histogram 
plt.figure(figsize=(5,4))
plt.hist(batting_first["win_by_runs"])
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.grid(True)  # Add grid for better readability
plt.show()
```

```python
#making the histogram 
plt.figure(figsize=(6,6))
plt.hist(batting_first["winner"], bins = 13)
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.grid(True)  # Add grid for better readability
plt.xticks(rotation=90)
plt.show()
```

```python
#making the bar plot for the tems with the most wins after batting first

batting_first["winner"].value_counts()
```

```python
#Getting the keys out because those are categorial values 
list(batting_first['winner'].value_counts().keys())
```

```python
#making the bar plot for the tems with the most wins after batting first

plt.figure(figsize=(4,5))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()), list(batting_first['winner'].value_counts()[0:3]), color=['yellow', 'green', 'pink'])
plt.xticks(rotation=90)
plt.show()
```

```python
data = list(batting_first['winner'].value_counts())
labels = list(batting_first['winner'].value_counts().keys())

# Explode each slice slightly
explode = [0.004] * len(data)

plt.figure(figsize=(20, 15))
plt.pie(data, labels=labels, autopct='%0.1f%%', textprops={'fontsize': 14}, explode=explode, pctdistance=1.05)
plt.show()
```
